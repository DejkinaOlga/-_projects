# Задание 1

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
 оценка 2 и меньше - низкий рейтинг
 оценка 4 и меньше - средний рейтинг
 оценка 4.5 и 5 - высокий рейтинг
Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
data = pd.read_csv('ratings.csv')
data.head(1)
```

```python
movies = pd.read_csv('movies.csv')
movies.head(1)
```

```python
# объединяем рейтинг и название
joined = data.merge(movies, on='movieId', how='left')
joined.head()
```

```python
# группируем по фильмам с подсчётом среднего рейтинга
joined.groupby('movieId').mean().round(1).reset_index()
```

```python
def class_(row):
    """классифицируем фильмы по рейтингу"""
    if row <= 2.0:
      return('Низкий рейтинг')
    elif row > 2.0 and row <= 4.0:
      return('Средний рейтинг')
    elif row == 4.5 or row == 5.0:
      return('Высокий рейтинг')
```

```python
# применяем метод apply к одному столбцу с созданием нового столбца
joined['class_'] = joined['rating'].apply(class_)
joined
```

```python
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
    }
```

```python
data = pd.read_csv('keywords.csv')
data.head()
```

```python
def contains_word(s, w):
  """проверяем наличие слова"""
  return f' {w} ' in f' {s} '
```

```python
def geo_class(row):
  """проверка с возвращением ключа"""
  for key, value in geo_data.items():
       for key, value in geo_data.items():
            for i in range(len(value)):

                if contains_word(row, value[i]) == True:

                    return(key)
  return('undefined')
```

```python
data['region'] = data['keyword'].apply(geo_class)
data
```

# Задание 3

Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл movies.csv и ratings.csv из базы grouplens.org.../movielens). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже. При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму: В переменную years запишите список из всех годов с 1950 по 2010. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм: для каждой строки пройдите по всем годам списка years если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’ Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

```python
years=list(range(1950,2011))
```

```python
print(years)
```

```python
def production_year(row):
  """выставляем год выпуска каждорй строке из названия фильма"""
  try:
        film_year = int(row[row.find('(')+1:row.find(')')])
        result = film_year
        if film_year not in years:
            result = 1900
        return result
  except:
        result = 1900
        return result
```

```python
# используем объединенный датафрейм из 1-го задания
joined['year'] = joined['title'].apply(production_year)
joined
```

```python
joined['year'].unique()
```

```python
# средний рейтинг по годам + сортировка

joined.groupby(['year']).mean().round(1).sort_values('rating', ascending=False).head(100)
```
