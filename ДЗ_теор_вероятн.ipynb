## Задание 3

Изучите 3 распределения из библиотеки scipy: Нормальное, Экспоненциальное, Стьюдента. Используя функцию плотности распредления (pdf), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения.

```python
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from pylab import rcParams
rcParams['figure.figsize']=10,6
```

### Нормальное распределение

```python
# создаём нормальное распределение
norm_rv_1=stats.norm(loc=2,scale=0.5)
norm_rv_2=stats.norm(loc=2,scale=1.5)
norm_rv_3=stats.norm(loc=3,scale=0.5)
norm_rv_4=stats.norm(loc=3,scale=1.5)
norm_rv_5=stats.norm(loc=6,scale=2)
```

```python
# строим график плотности вероятности для нормального распределения
x=np.linspace(0,8,100)
fig, ax = plt.subplots()
pdf_1=norm_rv_1.pdf(x)
pdf_2=norm_rv_2.pdf(x)
pdf_3=norm_rv_3.pdf(x)
pdf_4=norm_rv_4.pdf(x)
pdf_5=norm_rv_5.pdf(x)
plt.plot(x,pdf_1,'g',x,pdf_2,'r',x,pdf_3,'b',x,pdf_4,'c',x,pdf_5,'y')
plt.title('Графики плотности для нормального распределения')
plt.legend(['loc=2,scale=0.5', 'loc=2,scale=1.5', 'loc=3,scale=0.5', 'loc=3,scale=1.5', 'loc=6,scale=2'])
plt.show()
```

```python
# с изменением параметра loc (среднее значение) кривая нормального распределения смещается по оси х, без изменения формы
# с изменением параметра scale (стандартное отклонение) кривая изменяет форму: вытягивается при уменьшении scale и сплющивается при увеличении scale
```

### Экспоненциальное распределение

```python
# создаём экспоненциально распределение
e_rv_1=stats.expon(loc=2,scale=0.5)
e_rv_2=stats.expon(loc=2,scale=1.5)
e_rv_3=stats.expon(loc=3,scale=0.5)
e_rv_4=stats.expon(loc=3,scale=1.5)
e_rv_5=stats.expon(loc=6,scale=2)
```

```python
# строим график плотности вероятности для экспоненциального распределения

pdf_1=e_rv_1.pdf(x)
pdf_2=e_rv_2.pdf(x)
pdf_3=e_rv_3.pdf(x)
pdf_4=e_rv_4.pdf(x)
pdf_5=e_rv_5.pdf(x)
plt.plot(x,pdf_1,'g',x,pdf_2,'r',x,pdf_3,'b',x,pdf_4,'c',x,pdf_5,'y')
plt.title('Графики плотности для экспоненциального распределения')
plt.legend(['loc=2,scale=0.5', 'loc=2,scale=1.5', 'loc=3,scale=0.5', 'loc=3,scale=1.5', 'loc=6,scale=2'])
plt.show()
```

```python
# с изменением параметра loc (среднее значение) кривая смещается по оси х, без изменения формы
# с изменением параметра scale (стандартное отклонение) кривая изменяет форму: вытягивается при уменьшении scale и сплющивается при увеличении scale
```

### Распределение Стьюдента

```python
from scipy.stats import t
```

```python
# создаём экспоненциально распределение
t_rv_1=stats.t(loc=2, scale=0.5, df=1)
t_rv_2=stats.t(loc=2,scale=0.5, df=100)
t_rv_3=stats.t(loc=3,scale=0.5, df=100)
t_rv_4=stats.t(loc=3,scale=1, df=100)
t_rv_5=stats.t(loc=6,scale=2, df=100)
```

```python
# строим график плотности вероятности для распределения Стьюдента

pdf_1=t_rv_1.pdf(x)
pdf_2=t_rv_2.pdf(x)
pdf_3=t_rv_3.pdf(x)
pdf_4=t_rv_4.pdf(x)
pdf_5=t_rv_5.pdf(x)
plt.plot(x,pdf_1,'g',x,pdf_2,'r',x,pdf_3,'b',x,pdf_4,'c',x,pdf_5,'y')
plt.title('Графики плотности для распределения Стьюдента')
plt.legend(['loc=2, scale=0.5, df=1', 'loc=2,scale=0.5, df=100', 'loc=3,scale=0.5, df=100', 'loc=3,scale=1, df=100', 'loc=6,scale=2, df=100'])
plt.show()
```

```python
# с изменением параметра loc (среднее значение) кривая смещается по оси х, без изменения формы
# с изменением параметра scale (стандартное отклонение) кривая изменяет форму: вытягивается при уменьшении scale и сплющивается при увеличении scale
# с изменением параметра df (степень свободы): при увеличении степеней свободы приближается к нормальному распределению. При достаточно небольшом числе степеней свободы плотность распределения Стьюдента более пологая, чем нормальная кривая.
```
