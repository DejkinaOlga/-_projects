# Задание 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
data.head(1)
```

```python
# находим страны
filtered_countries = data.loc[ (data['country']=='Latvia') | (data['country']=='Lithuania') | (data['country']=='Estonia')]
filtered_countries.head(1)
```

```python
filtered_countries['country'].unique()
```

```python
filtered_countries['year'].unique()
```

```python
# задаем года, убираем отрицательные значения
yeas_countries=filtered_countries[(filtered_countries['year']>=2005)&(filtered_countries['year']<=2010)&(filtered_countries['quantity']>0)]
yeas_countries.head(1)
```

```python
yeas_countries['year'].unique()
```

```python
# задаём нужные категории
category_yeas_countries=yeas_countries[yeas_countries.category.isin([4,12,21])]
category_yeas_countries.head(1)
```

```python
category_yeas_countries['category'].unique()
```

```python
# считаем сумму
res_sum=category_yeas_countries['quantity'].sum()
print('Суммарное потребление по заданным условиям: ', res_sum)
```
