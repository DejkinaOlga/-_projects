## Задание 2. Вычислите градиент функции:6QCCExaG.jpg в точке (1, 2)

для функции двух переменных формула градиента:

Gradient-funktsii.png

Величина градиента функции, в которой производная имеет наибольшее значение определяется по формуле: для двух переменных:

Gradient-funktsii-formula.png

```python
import sympy as sym
```

```python
# задаём переменные
x,y=sym.symbols('x, y')
```

```python
# находим частные производные функции первого порядка в точке (1,2)
res_x=sym.diff(2*x**2*y**3+1/x+y**2*x+7,x).evalf(subs={x:1,y:2})
res_y=sym.diff(2*x**2*y**3+1/x+y**2*x+7,y).evalf(subs={x:1,y:2})
res_x, res_y
```

```python
# градиент функции в точке (1,2)
sym.sqrt(res_x**2+res_y**2)
```

## Задание 3. Постройте трехмерный график функции image.png

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import projections
```

```python
fig=plt.figure(figsize=(14,8))
ax=fig.gca(projection='3d')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
ax.set_title('Трехмерный график функции')

delta=0.025
x1=np.arange(-3,3,delta)
x2=np.arange(-3,3,delta)
x1,x2=np.meshgrid(x1,x2)
y=x2**2-x1**2*x2+2*x1*x2
surf=ax.plot_surface(x1,x2,y, cmap='hot', linewidth=0, antialiased=True)

```
