В этом задании нужно убедиться, что ЦПТ действительно работает.

Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n. Постройте гистрограммы средних этих выборок и сделайте выводы.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize']=20,12
```

```python
# задаём распределение
rv=sts.expon(loc=10,scale=1.5)
```

```python
# задаём выборку
sample = rv.rvs(size=1000)
```

```python
x = np.linspace(0,2,100)
pdf = rv.pdf(x) # график плотности
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples,f(x)')
plt.xlabel('x')
plt.show()
```

```python
# теперь из выборки выбираем несколько данных, усредним их и построим новое распределение
```

```python
n1 = 2                # Размеры выборок
n2 = 10
n3 = 50
n4 = 100
n5 = 500
a1 = np.array([])      # Массивы для сохранения средних по выборкам
a2 = np.array([])
a3 = np.array([])
a4 = np.array([])
a5 = np.array([])
samples_count = 1000  # Количество повторов генерации

for number in range(samples_count):
    sample1 = rv.rvs(size=n1)               # На каждом цикле генерируем выборки указанных размеров
    sample2 = rv.rvs(size=n2)
    sample3 = rv.rvs(size=n3)
    sample4 = rv.rvs(size=n4)
    sample5 = rv.rvs(size=n5)
    a1 = np.append(a1, sample1.mean())      # Считаем среднее по каждой выборке и добавляем в массив средних
    a2 = np.append(a2, sample2.mean())
    a3 = np.append(a3, sample3.mean())
    a4 = np.append(a4, sample4.mean())
    a5 = np.append(a5, sample5.mean())
```

```python
label_1 = 'sample of ' + str(n1) + ' histogram'   # Подписи для графиков
label_2 = 'sample of ' + str(n2) + ' histogram'
label_3 = 'sample of ' + str(n3) + ' histogram'
label_4 = 'sample of ' + str(n4) + ' histogram'
label_5 = 'sample of ' + str(n5) + ' histogram'
# YOUR CODE HERE
plt.hist(a1, density=True, label=label_1, alpha=0.5)         # Строим графики
plt.hist(a2, density=True, label=label_2, alpha=0.5)
plt.hist(a3, density=True, label=label_3, alpha=0.5)
plt.hist(a4, density=True, label=label_4, alpha=0.5)
plt.hist(a5, density=True, label=label_5, alpha=0.5)
# YOUR CODE HERE

plt.legend()
plt.ylabel('samples means')
plt.xlabel('x')
```

```python
# как видно из графиков, ЦПТ действительно работает. При этом, чем больше размер выборки (значение n), тем сильнее сходство гистограммы с нормальным распределением
```
